Usage of Makefiles have gone beyond original purpose of compilation and building of systems in C language.

Many people use them for running arbitrary commands like setup, linting, formatting, etc.,
in projects built on different languages and stacks, inlcuding Python.
For that purpose `Makefile` works fine, but has quite a lot of quirks related to its original purpose.
Recently a better substitute for `Makefile` for broader use cases appeared - it is called [Just](https://github.com/casey/just).


I've found that tool is very useful and decided to switch to it in all my projects.


## Makefile

But first let's recall how a Makefile is typically used.
This is an example of a Makefile I'm using for one of my projects:

```Makefile

.PHONY: *


activate:
	poetry shell

lint:
	poetry run ruff .

watchmedo:
	pip install watchdog

celery: watchmedo
	watchmedo auto-restart --directory=./ --pattern=*.py --recursive -- celery -A my-project worker  -l INFO

```
